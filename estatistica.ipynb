```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats

# Cria um DataFrame fictício para este exemplo
data = pd.DataFrame({'X': [1, 2, 3, 4, 5, 6],
                     'Y': [2, 3, 5, 4, 7, 8]})

# Realiza a regressão linear usando scipy.stats.linregress
slope, intercept, r_value, p_value, std_err = stats.linregress(data['X'], data['Y'])

# Calcula os valores de Y previstos com base na regressão
predicted_Y = slope * data['X'] + intercept

# Calcula os resíduos (diferença entre os valores observados e previstos)
residuals = data['Y'] - predicted_Y

# Usando lmplot para criar um gráfico de dispersão e uma linha de regressão
sns.lmplot(x='X', y='Y', data=data)

# Adicionando rótulos aos eixos e um título
plt.xlabel("Variável X")
plt.ylabel("Variável Y")
plt.title("Gráfico de Regressão Linear")

# Exibe os valores de a, b e r na legenda
plt.legend({'$a={:.2f}$'.format(slope), '$b={:.2f}$'.format(intercept), '$r={:.2f}$'.format(r_value)})

# Mostrando o gráfico de dispersão
plt.show()

# Criando um gráfico de resíduos
plt.figure()
plt.scatter(data['X'], residuals, c='r')
plt.axhline(y=0, color='b', linestyle='--')
plt.xlabel("Variável X")
plt.ylabel("Resíduos")
plt.title("Gráfico de Resíduos")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression

# Cria um DataFrame fictício para este exemplo
data = pd.DataFrame({'X': [1, 2, 3, 4, 5, 6],
                     'Y': [2, 3, 5, 4, 7, 8]})

# Separe as features (X) e o target (Y)
X = data[['X']]
Y = data['Y']

# Crie um modelo de regressão linear
model = LinearRegression()
model.fit(X, Y)

# Faça previsões com base no modelo
predicted_Y = model.predict(X)

# Calcule os resíduos (diferença entre os valores observados e previstos)
residuals = Y - predicted_Y

# Crie um gráfico de regressão linear
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(X, Y, label='Valores Observados')
plt.plot(X, predicted_Y, color='red', label='Regressão Linear')
plt.xlabel("Variável X")
plt.ylabel("Variável Y")
plt.title("Gráfico de Regressão Linear")
plt.legend()

# Crie um gráfico de resíduos
plt.subplot(1, 2, 2)
plt.scatter(X, residuals, c='r')
plt.axhline(y=0, color='b', linestyle='--')
plt.xlabel("Variável X")
plt.ylabel("Resíduos")
plt.title("Gráfico de Resíduos")

# Mostre os gráficos
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Defina os intervalos
bins = [0, 2, 3, 5, 7, 10, 15]

data = {
    'quantidade': [2.5, 4, 6, 8.5, 12.5],
    'urbano': [10, 40, 80, 50, 20],
    'rural': [30, 50, 15, 5, 0]
}

# Crie o DataFrame com base nos dados
df = pd.DataFrame(data)

# Crie o histograma
ax = sns.histplot(data=df, x="quantidade", weights="urbano", bins=bins, kde=False, color='blue', stat="percent")
for p in ax.patches:
    ax.annotate(f'{p.get_height():.1f}%', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom')
plt.title("Aluguéis urbanos")
plt.xlabel("Quantidade")
plt.ylabel("Percentagem de Urbano")
plt.show()

ax = sns.histplot(data=df, x="quantidade", weights="rural", bins=bins, kde=True, color='purple', stat='percent')
for p in ax.patches:
  ax.annotate(f'{p.get_height():.1f}%', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom')
plt.title("Aluguéis rurais")
plt.xlabel("Quantidade")
plt.ylabel("Percentagem de Rural")
plt.show()
```
